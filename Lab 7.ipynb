<h2>6). Напишіть функцію, яка приймає в рядок одне або декілька слів і повертає
той самий рядок, але зі зворотними всіма п’ятьма або більше буквами.</h2>

<h2>Рядки, передані, складатимуться лише з літер та пробілів.</h2>

<h2>Пробіли включатимуться лише тоді, коли є більше одного слова.</h2>

```python
def spinwords(sentence):
    return ' '.join([''.join(sorted(s,reverse=True)) if len(s)>=5 else s for s in sentence.split()])

print(spinwords("Hey fellow warriors"))
print(spinwords("This is a test"))
print(spinwords("This is another test"))
```

<h2>7). Проект: Створити застосунок для кодування та розкодування за допомогою коду
Цезаря. Розробити можливість вибору напряму кодування (кодування чи
розкодування), вводу з клавіатури речень та зсуву.</h2>

```python
llst = ['а','б','в','г','д','е','ж','з','и','й','к','л','м','н','о','п','р','с','т','у','ф','х','ц','ч','ш','щ','ъ','ы','ь','э','ю','я']
blst = ['А','Б','В','Г','Д','Е','Ж','З','И','Й','К','Л','М','Н','О','П','Р','С','Т','У','Ф','Х','Ц','Ч','Ш','Щ','Ъ','Ы','Ь','Э','Ю','Я']
 
 
def encryptCaesar(msg, shift=3):
    ret = ""
    for x in msg:
        if x in llst:
            ind = llst.index(x)
            ret += llst[ind+shift]
        elif x in blst:
            ind = blst.index(x)
            ret += blst[ind+shift]
        else:
            ret += x
    return ret
 
def decryptCaesar(msg, shift=3):
    ret = ""
    for x in msg:
        if x in llst:
            ind = llst.index(x)
            ret += llst[ind-shift]
        elif x in blst:
            ind = blst.index(x)
            ret += blst[ind-shift]
        else:
            ret += x
    return ret
 
print(encryptCaesar("Да здравствует салат Цезарь!"))
print(decryptCaesar("Зг кзугефхецих фгогх Щикгуя!"))
```
